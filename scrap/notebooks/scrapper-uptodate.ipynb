**Step 1:** Get articles from [UpToDate](https://www.uptodate.com/contents/coronavirus-disease-2019-covid-19).
We can use the requests library to do this.

```python
# import statements
import requests
from bs4 import BeautifulSoup
```

```python
# fetch web page
r = requests.get("https://www.uptodate.com/services/app/contents/search/2/json?index=0~10&language=en&max=10&search=covid+19")
```

This website, however redirects to a single, complete article. There's links to other ones, however. Notice the ending path arg.

**Step 2:** Use requests to get json of the API

```python
articles = r.json()['data']['searchResults']
```

```python
articles[0]['url'].split('/')[2].split('?')[0]
```

```python
def get_abstract(url):
    url = url.split('/')[2].split('?')[0]
    d = requests.get("https://www.uptodate.com/services/app/contents/topic/" + url + "/json")
    content_soup = BeautifulSoup(d.json()["data"]['bodyHtml'], "html.parser")
    print(content_soup.select_one("#topicText").get_text())
```

```python
get_abstract(articles[0]['url'])
```
